```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D #mpl toolkits to visualize the raster data 
file_path = "C:\\Users\\megha\\Desktop\\.ipynb_checkpoints\\myfilefordemo.TIF"


with rasterio.open(file_path) as src:
   
    band1 = src.read(1)
    
    x_min, y_max = src.transform * (0, 0)
    x_max, y_min = src.transform * (src.width, src.height)
    x = np.linspace(x_min, x_max, src.width)
    y = np.linspace(y_min, y_max, src.height)
    X, Y = np.meshgrid(x, y)
    
    fig = plt.figure() # Plotting
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, band1, cmap='viridis')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Pixel Value')
    ax.set_title('3D Projection of Raster File')
    plt.show()
```
